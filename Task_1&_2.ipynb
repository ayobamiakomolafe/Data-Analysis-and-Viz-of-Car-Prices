```python
"Extract dataset to show the cheapest and expensive cars based on fuel_type and system"

df1=df.sort_values('price').tail(100)[['fuel_type', 'fuel_system']].reset_index().drop('index', axis=1)
df1=df1.groupby(['fuel_type', 'fuel_system']).size().to_frame().reset_index()
df1['fuel_type/system']=df1['fuel_type'] + '_'+ df1['fuel_system'] 
df1=df1.rename(columns={0:'Expensive_100'})

df2=df.sort_values('price').head(100)[['fuel_type', 'fuel_system']].reset_index().drop('index', axis=1)
df2=df2.groupby(['fuel_type', 'fuel_system']).size().to_frame().reset_index()
df2['fuel_type/system']=df2['fuel_type'] + '_'+ df2['fuel_system'] 
df2=df2.rename(columns={0:'Cheapest_100'})

df2=df2[['Cheapest_100','fuel_type/system']]
df1=df1[['Expensive_100',	'fuel_type/system']]

df3=df1.merge(df2, how='outer', on='fuel_type/system').fillna(0).set_index('fuel_type/system')
df3['Cheapest_100']=0-df3['Cheapest_100']


font_color = '#525252'
hfont = {'fontname':'Calibri'}
facecolor = '#eaeaf2'
color_red = '#fd625e'
color_blue = '#01b8aa'
index = df3.index
title0 = 'Expensive_100'
title1 = 'Cheapest_100'

fig, axes = plt.subplots(figsize=(10,5), facecolor=facecolor, ncols=2, sharey=True)
fig.tight_layout()

axes[0].barh(index, df3['Expensive_100'], align='center', color=color_red, zorder=10)
axes[0].set_title(title0, fontsize=18, pad=15, color=color_red, **hfont)
axes[1].barh(index, df3['Cheapest_100'], align='center', color=color_blue, zorder=10)
axes[1].set_title(title1, fontsize=18, pad=15, color=color_blue, **hfont)
axes[0].invert_xaxis() 
axes[1].invert_xaxis() 


plt.savefig("exp.png", bbox_inches='tight')
files.download("exp.png") 
```

# **Model** **Development**

## **Data Cleaning**

```python
df.drop('car_company', axis=1, inplace=True)
```

```python
df.drop('Car_ID', axis=1, inplace=True)
```

```python

nu=df['fuel_type'].nunique()
num=range(nu)
uni=df['fuel_type'].unique()
for a,b in zip(num,uni):
    df['fuel_type'].replace(b,a, inplace=True)
    

nu=df['aspiration'].nunique()
num=range(nu)
uni=df['aspiration'].unique()
for a,b in zip(num,uni):
    df['aspiration'].replace(b,a, inplace=True)

nu=df['door_number'].nunique()
num=range(nu)
uni=df['door_number'].unique()
for a,b in zip(num,uni):
    df['door_number'].replace(b,a, inplace=True)

nu=df['car_body'].nunique()
num=range(nu)
uni=df['car_body'].unique()
for a,b in zip(num,uni):
    df['car_body'].replace(b,a, inplace=True)

nu=df['drive_wheel'].nunique()
num=range(nu)
uni=df['drive_wheel'].unique()
for a,b in zip(num,uni):
    df['drive_wheel'].replace(b,a, inplace=True)

nu=df['engine_location'].nunique()
num=range(nu)
uni=df['engine_location'].unique()
for a,b in zip(num,uni):
    df['engine_location'].replace(b,a, inplace=True)

nu=df['engine_type'].nunique()
num=range(nu)
uni=df['engine_type'].unique()
for a,b in zip(num,uni):
    df['engine_type'].replace(b,a, inplace=True)


nu=df['fuel_system'].nunique()
num=range(nu)
uni=df['fuel_system'].unique()
for a,b in zip(num,uni):
    df['fuel_system'].replace(b,a, inplace=True)

nu=df['company'].nunique()
num=range(nu)
uni=df['company'].unique()
for a,b in zip(num,uni):
    df['company'].replace(b,a, inplace=True)
```

```python
df.info()
```

```python
X=df.drop('price', axis=1)  
Y=df['price']                
```

```python
"split data to train and test set"
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
"Random Forest"

from sklearn.ensemble import RandomForestRegressor
Rf= RandomForestRegressor()
Rf.fit(X_train, y_train)
Rf.score(X_test, y_test)
```

```python
plt.barh(X.columns, Rf.feature_importances_, color='purple')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Bar graph of features against their importance to price value')
plt.savefig("exp.png")
files.download("exp.png")
```

```python
"linear regression"

from sklearn.linear_model import LinearRegression
Lr= LinearRegression()
Lr.fit(X_train, y_train)
Lr.score(X_test, y_test)
```

```python
"Coefficient values"
df_coef=pd.DataFrame(Lr.coef_, X.columns).reset_index()
df_coef.rename(columns={"index": 'Column', 0:'Coefficients'})
```
